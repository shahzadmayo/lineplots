# Draw the LinePlot using the Different parameter and diffrent DataFrame

```python
# import important libabry 

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt


#load_data set
phool= sns.load_dataset('iris')
phool.head()
```

```python
# Simple Draw the line Plot
sns.lineplot(x='petal_length',y='sepal_width', data=phool)

# adding the limit
#plt.xlim(2)
#plt.ylim(1)
plt.show()
```

```python
# Add Some new feature in line plot
sns.lineplot(x='petal_length',y='sepal_width', data=phool, hue='species', style='species')
plt.title('Base on the species',color='white')
plt.show()
```

## Now Using the Different Dataframe and draw the lineplot

```python
# loading dataset
data = sns.load_dataset("tips")
  
# draw lineplot
# hue by sex
# style to hue
sns.lineplot(x="total_bill", y="size",
             hue="sex", style="sex",
             data=data)
plt.title('Line Plot gender Base', color='white')
plt.show()
```

```python
fmri = sns.load_dataset("fmri")
# plotting lineplot for signal with respect to timepoint
# using relplot, kind = "line"
# code added by devanshigupta1304
sns.lineplot( x = "timepoint",
             y = "signal",
             data = fmri)
fmri.head()
```

## Creating multiple line plot using hue parameters.

```python

sns.lineplot( x = "timepoint",
             y = "signal",
             hue = "region",
             data = fmri)
```

## Creating multiple line plots using style parameters.

```python
sns.lineplot( x = "timepoint",
             y = "signal",
             hue = "region",
             style = "event",
             data = fmri)
```

## Creating multiple line plots using size parameters. We can use size attributes to change the line pattern.

```python
sns.lineplot( x = "timepoint",
             y = "signal",
             size = "event",
             data = fmri)
```

## Grouping data points on the basis of category, here as region and event. We can add them as hue and style semantics. This will change the line pattern by default.

```python
# using relplot with kind = "line"
# adding style and hue semantic
sns.relplot( x = "timepoint",
            y = "signal",
            hue = "region", style = "event",
            kind = "line", data = fmri)
plt.title('Relplot usnig kind line ', color='white')
```

## We can also represent standard deviation instead of confidence interval at each time point, for a large dataset, as it can be very time-consuming

```python
# plotting lineplot for signal with respect to timepoint
# using relplot, kind = "line"
# using standard deviation instead of confidence interval
sns.relplot(x = "timepoint", y = "signal",
            kind = "line", ci = "sd",
            data = fmri)
```

## We can use style and hue with different columns in a data set.

```python
#using lineplot with style = column_name
# adding style and hue semantic
sns.lineplot(x = "timepoint", y = "signal",
             hue = "subject",style = "event",
             data = fmri)
```

## Using palette attributes

```python
# using lineplot
# adding palette, style and hue semantic
sns.lineplot( x = "timepoint", y = "signal",
             hue = "subject", style = "event",
             palette = "YlOrRd_r", data = fmri)
```

## Adding error bar in line plot using err_style attributes.

```python
sns.lineplot( x = "timepoint",
             y = "signal",
             data = fmri,
             err_style="bars")
```
